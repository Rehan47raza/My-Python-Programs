```python
#Create tkinter application to accept user name in a text widget.

from tkinter import *
from tkinter import ttk

win= Tk()

win.geometry("250x250")

def display_text():
    global entry
    string= entry.get()
    label.configure(text=string)


label=Label(win, text=" ", font=("Courier 22 bold"))
label.pack()


entry= Entry(win, width= 40)
entry.focus_set()
entry.pack()


ttk.Button(win, text= "Okay",width= 20, command= display_text).pack(pady=20)

win.mainloop()



```

```python
import tkinter as tk
import math

def generate():
    try:
        result = float(num1.get())*float(num1.get())*math.pi
    except Exception as ex:
        print(ex)
        result = 'error'

    num3.set(result)

root = tk.Tk()
root.geometry("250x250")

num1 = tk.StringVar()
num3 = tk.StringVar()

tk.Label(root, text="Radius of Circle:").grid(row=0, column=0)
tk.Label(root, text="Area of Circle:").grid(row=2, column=0)

tk.Entry(root, textvariable=num1).grid(row=0, column=1)
tk.Entry(root, textvariable=num3).grid(row=2, column=1)

button = tk.Button(root, text="Calculate", command=generate)
button.grid(row=3, column=1)

root.mainloop()
```

```python
#Create a tkinter application to display a menu with three items

from tkinter import *

win=Tk()
def hello():
    print('Hello World!!')
    
def Save():
    print('File Saved Successfully')

menubar=Menu(win)
filemenu=Menu(win)
filemenu.add_command(label='open',command=hello)
filemenu.add_separator()
filemenu.add_command(label='Save',command=Save)
filemenu.add_separator()
filemenu.add_command(label='Exit',command=win.destroy)
filemenu.add_separator()


menubar.add_cascade(label='File',menu=filemenu)
win.config(menu=menubar)
win.mainloop()



```

```python
#Create a tkinter application to display a radio button group with three items

from tkinter import *
win=Tk()
def Sel():
    selection="You selected the option"+str(var.get())
    label.config(text=selection)
    
    
var=IntVar()
R1=Radiobutton(win,text="option1",variable=var,value=1,command=Sel)
R1.pack()

R2=Radiobutton(win,text="option2",variable=var,value=2,command=Sel)
R2.pack()

R3=Radiobutton(win,text="option3",variable=var,value=3,command=Sel)
R3.pack()

label=Label(win)
label.pack()
win.mainloop()

```

```python
#Create a tkinter application to display three check boxes.

from tkinter import *

def Sel1():
    if(CheckVar1.get()==1):
        selection='You selected Python programming option:'
        
    else:
        selection='You deselected the python programming options:'
        
    label.config(text=selection)

    
    
    
def Sel2():
    if(CheckVar2.get()==1):
        selection='You selected Tkinter option:'
        
    else:
        selection='You deselected Tkinter options:'
        
    label.config(text=selection)
    
    
def Sel3():
    if(CheckVar3.get()==1):
        selection='You selected Skilllync option:'
        
    else:
        selection='You deselected Skilllync options:'
        
    label.config(text=selection)
    

win=Tk()
CheckVar1=IntVar()
CheckVar2=IntVar()
CheckVar3=IntVar()

c1=Checkbutton(win,text="python programming",variable=CheckVar1,onvalue=1,offvalue=0,height=5,width=20,command=Sel1)

c2=Checkbutton(win,text="Tkinter Library",variable=CheckVar2,onvalue=1,offvalue=0,height=5,width=20,command=Sel2)

c3=Checkbutton(win,text="Skill lync ",variable=CheckVar3,onvalue=1,offvalue=0,height=5,width=20,command=Sel3)

c1.pack()
c2.pack()
c3.pack()
label=Label(win)
label.pack()
win.mainloop()




    
    
```

```python
#Create a tkinter application which contains a list box with five items.

from tkinter import *

def my_upd():
    print(l1.get(ACTIVE))

win=Tk()
l1=Listbox(win,height=6)
l1.grid(row=1,column=1)
my_list=['tkinter','python','Numpy','Skilllync','Hello']

for elements in my_list:
    l1.insert(END,elements)
    
b1=Button(win,text='Show',width=10,bg='Yellow',command=my_upd)
b1.grid(row=1,column=2)
win.mainloop()



```
