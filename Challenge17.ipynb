```python
#class called number which contains an integer. It should contain various methods to set the value of the integer,

class Number():
    
    def __init__(self, n=0):
        self.num = n
        
    def get_value(self):
        return self.num
    
    def set_value(self):
        self.num=n
        
    def check_value(self):
        
        if self.num>0:
            print('Entered number is positive')
        else:
            print ('Entered number is negative')
    
    def print_value(self):
        print('The value entered is :',self.num)


my_num= int(input("Enter the number..."))
my_instance=Number(my_num)
print(my_instance.get_value())
print(my_instance.print_value())
print(my_instance.check_value())


```

```python
#class that can calculate the perimeter/ circumference and area of a regular shape.

import math

class circle_compute():
    
    def __init__(self,my_radius):
        self.radius=my_radius
        
    def area_calculate(self):
        return math.pi*(self.radius**2)
    
    def perimeter_calculate(self):
        return 2*math.pi*self.radius
    
my_result = float(input("Enter the radius of circle..."))

my_instance = circle_compute(my_result)


print("The Area of circle is :")

print(my_instance.area_calculate())

print("The perimeter of circle is :")

print(my_instance.perimeter_calculate())
```

```python
#Create a class called Date. It should contain data members day, month and year

from datetime import datetime
class Date():
    
    def __init__(self,y=0,m=0,d=0):
        self.year=y
        self.month=m
        self.day=d
        
    def set_date(self):
        self.year=y
        self.month=m
        self.day=d
        
        
    def display_date(self):
        return self.year,self.month,self.day
    
    def __eq__(self,other):
        if self.year==other.year and self.month==other.month and self.day==other.day:
            return True
        else:
            return False
        

d1=Date(2022,6,12)
d3=Date(2022,6,12)
d2=Date(2022,5,12)
print('Date 1:',d1.display_date())
print('Date 2:',d2.display_date())
print('Date 3:',d3.display_date())

print(d1==d3)
print(d2==d3)
    
        


```

```python
#Create a class called complex containing real and imaginary parts

class complex:
    def __init__(self, a, b):
        self.a = a
        self.b = b
 
    def __add__(self, other):
        return self.a + other.a, self.b + other.b
    
    def __eq__(self,other):
        if self.a==other.a and self.b==other.b:
            return True
        else:
            return False
        
    def __sub__(self, other):
        return self.a - other.a, self.b - other.b
        
    
Ob1 = complex(6, 5)
Ob2 = complex(2, 3)
Ob3 = complex(2, 3)
Ob4 = complex(3, 4)

#checking the object type

Ob5=Ob1+Ob2
Ob6=Ob4-Ob3

print(Ob5==Ob6)

print(Ob1 is Ob2) # checking the pointing to same object

print(Ob2==Ob3) # Checking the attributes

```
